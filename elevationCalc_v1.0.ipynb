```python
# Goes with GBO/Phase 2 - Spillover Correction.
# Written by Alyssa Bulatek.
# Version 1.0 (estimated)

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rc
import os

plt.rc('text', usetex = True) # Use LaTeX font in plots
plt.rcParams['text.latex.preamble'] = [r'\usepackage{gensymb}',r'\usepackage{sansmath}',r'\sansmath']
```

```python
######## Edit changing parameters here ########

elev_angle = '60'
cut = '45'
norm = True

###############################################

if norm == True:
    spec = '_norm_'+cut
if norm == False:
    spec = '_reg_'+cut
    
if cut == '0':
    date_prefix = 'June19_Die_1p1_4_E'
if cut == '45':
    date_prefix = 'June16_Die_1p1_4_D'
if cut == '90':
    date_prefix = 'June19_Die_1p1_4_H'

# Create the filename and filepath given the inputs
filename = date_prefix+'_ab_efficiencies'+spec+'_elev_'+elev_angle+'.csv'
filepath = os.getcwd()+'/Phase 2.1 - Spillover Efficiency Comparison/'+date_prefix+'/'+'/'+filename
```

```python
# Extract data from the file
df = pd.read_csv(filepath)
freq = df['Freq'].values
E_sp = df['Spillover_dish'].values
E_sp_b = df['Spillover_backlobe'].values
```

```python
# Plot spillover efficiency as a function of frequency
plt.figure(dpi = 400)
plt.plot(freq,E_sp,label=r"$e_{sp,d}$",color='xkcd:azure')
plt.plot(freq,E_sp_b,label=r"$e_{sp,b}$",color='xkcd:salmon',linestyle='--')
plt.legend(loc="lower left")
plt.xlabel("Frequency (MHz)")
plt.ylabel("Spillover efficiency")
plt.xlim(1000,4100)
plt.ylim(0,1)
if norm == True:
    plt.title("Spillover Efficiency, cut angle = "+cut+"$\degree$")
else:
    plt.title("Spillover Efficiency, cut angle = "+cut+"$\degree$, non-normalized")
plt.savefig(os.getcwd()+'/Phase 2.1 - Spillover Efficiency Comparison/'+date_prefix+'/ab'+spec+'.png', transparent=True)
plt.plot()
```
