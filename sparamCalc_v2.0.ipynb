```python
# Goes with GBO/Phase 6 - S11 Parameters.
# Written by Alyssa Bulatek.
# Version 2.0 (estimated)

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rc
import os

plt.rc('text', usetex = True) # Use LaTeX font in plots
plt.rcParams['text.latex.preamble'] = [r'\usepackage{gensymb}']
#                                       r'\usepackage{sansmath}',
#                                       r'\sansmath']
```

```python
######## Edit changing parameters here ########

filename = 'complete_s11_data_poster_'

nwhi = 'no_window_hybrid_ideal.csv'
nwhr = 'no_window_hybrid_real.csv'
whi = 'window_hybrid_ideal.csv'
whr = 'window_hybrid_real.csv'

###############################################

# Create the filepath given the inputs

filepath_nwhi = os.getcwd()+'/'+filename+nwhi
filepath_nwhr = os.getcwd()+'/'+filename+nwhr
filepath_whi = os.getcwd()+'/'+filename+whi
filepath_whr = os.getcwd()+'/'+filename+whr
```

```python
# Extract data from the files
nwhi = pd.read_csv(filepath_nwhi)
nwhr = pd.read_csv(filepath_nwhr)
whi = pd.read_csv(filepath_whi)
whr = pd.read_csv(filepath_whr)

nwhi_freq = nwhi['Frequency[GHz]'].values
nwhr_freq = nwhr['Frequency[GHz]'].values
whi_freq = whi['Frequency[GHz]'].values
whr_freq = whr['Frequency[GHz]'].values

nwhi_S11 = nwhi['S11[dB]'].values
nwhr_S11 = nwhr['S11[dB]'].values
whi_S22 = whi['S22[dB]'].values
whr_S11 = whr['S11[dB]'].values
```

```python
# Plot S11 parameters as a function of frequency, no window, hybrid
plt.figure(dpi = 600)
plt.rc('font', family='serif')
plt.plot(nwhi_freq,nwhi_S11,color="xkcd:sea blue",label='ideal hybrid')
plt.plot(nwhr_freq,nwhr_S11,color="xkcd:orange",label='real hybrid')
plt.xlabel("Frequency (GHz)", size = 14)
plt.ylabel(r"S\textsubscript{11} (dB)", size = 14)
plt.xlim(0.7,4.2)
plt.xticks(np.arange(0.7,4.3,0.5))
plt.ylim(-65,-6)
plt.legend(loc="lower left")
plt.title("UWB Receiver Input Reflection, no window", size = 16)
plt.grid()
plt.savefig("/Users/alyssabulatek/Desktop/S11_no_window.png", transparent=False, facecolor='none', edgecolor='none')
plt.plot()
```

```python
# Plot S11 parameters as a function of frequency, yes window, hybrid
plt.figure(dpi = 600)
plt.rc('font', family='serif')
plt.plot(whi_freq,whi_S22,color="xkcd:sea blue",label='ideal hybrid')
plt.plot(whr_freq,whr_S11,color="xkcd:orange",label='real hybrid')
plt.xlabel("Frequency (GHz)", size = 14)
plt.ylabel(r"S\textsubscript{11} (dB)", size = 14)
plt.xlim(0.7,4.2)
plt.xticks(np.arange(0.7,4.3,0.5))
plt.ylim(-65,-6)
plt.legend(loc="lower left")
plt.title("UWB Receiver Input Reflection, window", size = 16)
plt.grid()
plt.savefig("/Users/alyssabulatek/Desktop/S11_window.png", transparent=False, facecolor='none', edgecolor='none')
plt.plot()
```
