```python
# Goes with Phase B - Mode Calculations.
# Written by Alyssa Bulatek.
# Version 1.0

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os

plt.rc('text', usetex = True) # Use LaTeX font in plots
plt.rcParams['text.latex.preamble'] = [r'\usepackage{gensymb}']
#                                       r'\usepackage{sansmath}',
#                                       r'\sansmath']

# Set user-specified parameters.
path = "/Users/alyssabulatek/Desktop/Mode_calculations/Trial on March 31 (final)/"
filename = "Tef_trial_march_31.csv"

# Extract coefficients from file.
filepath = path+"/"+filename
df = pd.read_csv(filepath)
freq = df['Frequency (GHz)'].values
TE11_percent = df['%TE11'].values
TM11_percent = df['%TM11'].values
TE12_percent = df['%TE12'].values
TM12_percent = df['%TM12'].values
TE13_percent = df['%TE13'].values
TM13_percent = df['%TM13'].values

# Normalize mode coefficients to the largest coefficient.
#coeff = coeff/max(coeff)
# Rewrite as a percentage.
#coeff_percent = coeff*100

# Plot mode coefficient as a function of frequency.
plt.figure(dpi = 600)
plt.rc('font', family='serif')
plt.plot(freq,TE11_percent,color="xkcd:sky blue",marker='.',label="TE11")
plt.plot(freq,TM11_percent,color="xkcd:orange",marker='.',label="TM11")
plt.plot(freq,TE12_percent,color="xkcd:grey",marker='.',label="TE12")
plt.plot(freq,TM12_percent,color="xkcd:goldenrod",marker='.',label="TM12")
plt.plot(freq,TE13_percent,color="xkcd:blue",marker='.',label="TE13")
plt.plot(freq,TM13_percent,color="xkcd:green",marker='.',label="TM13")
plt.xlabel("Frequency (GHz)", size = 14)
plt.ylabel("Mode coefficient contribution (\%)", size = 14)
plt.xlim(0.7,4.2)
plt.xticks(np.arange(0.7,4.3,0.5))
plt.ylim(0,100)
plt.title("UWB Receiver Mode Coefficients, Model D (Teflon)", size = 16)
plt.legend(loc="upper right")
plt.grid()
plt.savefig("/Users/alyssabulatek/Desktop/mode_coeff_march28.png", transparent=False)
            #, facecolor='none', edgecolor='none')
plt.plot()
```
