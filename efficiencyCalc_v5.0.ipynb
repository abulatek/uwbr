```python
# Goes with GBO/Phase 1 - Efficiencies.
# Written by Alyssa Bulatek.
# Version 5.0 (estimated)

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rc
import os

plt.rc('text', usetex = True) # Use LaTeX font in plots
plt.rcParams['text.latex.preamble'] = [r'\usepackage{gensymb}']
#                                       r'\usepackage{sansmath}',
#                                       r'\sansmath']

######## Edit changing parameters here ########

subfolder = 'July_Final_H'
cut = '90'

path = '/Users/alyssabulatek/Desktop/GBO/Phase 1 - Efficiencies/Phase 1.3 - Final Final Efficiencies/'
filename = subfolder+'_ab_efficiencies_norm_'+cut+'.csv'

###############################################

#if cut == '0':
#    subfolder = 'E'
#if cut == '45':
#    subfolder = 'D'
#if cut == '90':
#    subfolder = 'H'

# Create the filename and filepath given the inputs
filepath = path+filename

# Extract data from the file
df = pd.read_csv(filepath)
freq = df['Freq'].values
E_sp = df['Spillover'].values
E_ill = df['Illumination'].values
if cut == '45':
    E_xp = df['XPolarization'].values
E_ph = df['Phase'].values
E_tot = df['Total'].values

# Plot individual efficiencies and total efficiency as a function of frequency on the same plot
plt.figure(dpi = 600)
plt.rc('font', family='serif')
plt.plot(freq/1000,E_sp,label=r"$e_{sp}$",color='xkcd:azure',linestyle="--")
plt.plot(freq/1000,E_ill,label=r"$e_{ill}$",color='xkcd:orange',linestyle=":")
plt.plot(freq/1000,E_ph,label=r"$e_{ph}$",color='xkcd:teal',linestyle="-.")
if cut == '45':
    plt.plot(freq/1000,E_xp,label=r"$e_{xp}$",color='xkcd:magenta',linestyle=(0,(3,1,1,1,1,1)))
plt.plot(freq/1000,E_tot,label=r"$e_{tot}$",color='xkcd:black')
plt.legend(loc="lower left")
plt.xlabel("Frequency (GHz)", size = 14)
plt.ylabel("Efficiency", size = 14)
plt.xlim(0.7,4.2)
plt.xticks(np.arange(0.7,4.3,0.5))
plt.ylim(0,1)
plt.grid()
plt.title("UWB Receiver Efficiencies, Model A, cut angle = "+cut+"$\degree$", size = 14)
plt.savefig("/Users/alyssabulatek/Desktop/"+subfolder+"_efficiency"+cut+".png", transparent=False)
#, facecolor='none', edgecolor='none')
plt.plot()
```
